# Literature Network

```python
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
%pylab inline
import seaborn 
```

```python
G = nx.read_gml('literature-network.gml')
```

```python
G.nodes()
```

```python
G.add_edges_from([('Aummann 1981', 'Neyaman 1985'), ('Neyaman 1985', 'Dilip Rubinstein 1988'),
                  ('Dilip Rubinstein 1988', 'Kalai 1988'), ('Kalai 1988', 'Fogel 1994'), ('Fogel 1994', 'Miller 1994'),
                  ('Miller 1994', 'Ashlock & Wife 1995'), ('Ashlock & Wife 1995', 'Engineering 2011')])
dotted = [('Kalai 1988', 'Aummann 1981'), ('Kalai 1988', 'Neyaman 1985'), ('Dilip Rubinstein 1988', 'Rubinstein 1986'),
          ('Dilip Rubinstein 1988', 'Ben-Porath 1986'), ('Dilip Rubinstein 1988', 'Megiddo 1986'),
          ('Neyaman 1985', 'Ben-Porath 1986'), ('Neyaman 1985', 'Megiddo 1986')]
```

```python
# draw 
fig = plt.figure(figsize=(50, 50))

plt.title("Literature Network", fontsize=50)
pos = nx.spring_layout(G)
nx.draw_networkx_edges(G, pos, edgelist=dotted, style='dashed')
network = nx.draw_networkx(G, pos, node_color='b', width=2)
plt.show()
```
