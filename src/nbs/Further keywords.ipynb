```python
import pandas as pd

import numpy as np
import collections
import re
from itertools import islice
```

```python
auction = pd.read_csv('../data/auction_theory_articles_meta_data.csv', index_col=0)
```

```python
price = pd.read_csv('../data/price_of_anarchy_articles_meta_data.csv')
```

```python
def data_to_text(df):
    text = df[['abstract', 'title', 'primary_category', 'category']].values
    flat_text =  [item for sublist in text for item in sublist]
    flat_text = [item for item in flat_text if type(item) is str]
    
    return ' '.join(flat_text)
```

```python
text = data_to_text(auction)
```

```python
words = re.findall("\w+", 
   text)
count = collections.Counter(zip(words, islice(words, 1, None)))
```

```python
count.most_common(1000)
```
