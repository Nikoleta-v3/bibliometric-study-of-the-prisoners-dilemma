```python
import pandas as pd
```

```python
df = pd.read_json('../data/data_nov_2017.json')
```

```python
df.head()
```

```python
def check_articles_provencance_on_keyword(key, df, key_in='abstract'):
    
    temp = df[df[key_in].str.contains(key, case=False, na=False)]
    
    flag_in_title = temp[temp['title'].str.contains("prisoner's dilemma", case=False, na=False)]
    flag_in_abstract = temp[temp['abstract'].str.contains("prisoner's dilemma", case=False, na=False)]       
    return len(temp['title'].unique()), len(flag_in_title['title'].unique()), len(flag_in_abstract['title'].unique())
```

```python
keywords = ['prisoners evolution', 'R Axelrod', 'memory one strategy', 'tit-for-tat',
            'tit for tat', 'zero determinant strategies', 'prisoner game theory']
```

```python
checks = []
for key in keywords:
    checks.append(check_articles_provencance_on_keyword(key, df))
```

```python
pd.DataFrame(checks, columns=['$N_A$ collected based on key', 
                              "$N_A$ prisoner's in title", 
                              "$N_A$ prisoner's in abstract"])
```
