```python
import networkx as nx

import matplotlib.pyplot as plt
```

```python
graphs = []
for filename in ["auction_graph.gml", "anarchy_graph.gml"]:
    graphs.append(nx.read_gml("../data/networks/{}".format(filename)))
```

```python
clusters = [max(nx.connected_component_subgraphs(G), key=len) for G in graphs]
```

```python
for G, label in zip(
    clusters, ["auction_network_cluster.pdf", "anarchy_network_cluster.pdf"]
):
    pos = nx.spring_layout(G)
    fig, axes = plt.subplots(figsize=(15, 15))
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color="black")
    nx.draw_networkx_edges(G, pos, alpha=0.5, edge_color="grey")

    limits = plt.axis("off")

    plt.savefig("../../assets/{}".format(label), transparent=True, bbox_inches="tight")
```
